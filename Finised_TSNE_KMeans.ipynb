```python
import numpy as np
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = load_digits().data
y = load_digits().target
```

```python
tsne = TSNE()
```

```python
X_2d = tsne.fit_transform(X)
```

```python
colors = ['r', 'g', 'b', 'c', 'm', 'y', 'k', 'brown', 'orange', 'purple']
```

```python
for i, c in zip(np.arange(10), colors):
    plt.scatter(X_2d[y == i, 0], X_2d[y == i, 1], c=c, label=i)
plt.legend()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=10, n_init='auto', max_iter=1000,
                random_state=42)
```

```python
kmeans.fit(X)
```

```python
y_pred = kmeans.predict(X)
```

```python
for i, c in zip(np.arange(10), colors):
    plt.scatter(X_2d[y_pred == i, 0], X_2d[y_pred == i, 1], c=c, label=i)
plt.legend()
```
