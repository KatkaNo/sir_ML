```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('weatherAUS.csv')
df
```

```python
df['RainTomorrow'].unique()
```

```python
df['RainTomorrow'] = df['RainTomorrow'].map({'Yes':1, 'No':0})
df['RainToday'] = df['RainToday'].map({'Yes':1, 'No':0})

df = df[df['RainTomorrow'] == df['RainTomorrow']]
df['RainTomorrow'].unique()
```

```python
df['RainTomorrow'].sum()/df.shape[0]
```

```python
y = df.RainTomorrow
x = df.drop(columns=['RainTomorrow'])

x = x.select_dtypes([np.number])
x = x.fillna(x.median())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver='lbfgs', max_iter = 10000)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

(y_pred == y_test).sum() / len(y_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix

conf_matrix = confusion_matrix(y_test, y_pred)
conf_matrix
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

sns.heatmap(conf_matrix, annot=True, fmt='.0f', linewidths=0.5, cmap='Blues_r')
plt.ylabel('true label')
plt.xlabel('Predicted label')
```

```python
TP = conf_matrix[1,1] #классификатор верно отнёс объект к рассматриваемому классу (дождь ожидался и был)
FN = conf_matrix[1,0] #классификатор неверно утверждает, что объект не принадлежит к рассматриваемому классу (дождь не ожидался, но был)
FP = conf_matrix[0,1] #классификатор неверно отнёс объект к рассматриваемому классу (дождь ожидался, но не было)
TN = conf_matrix[0,0] #классификатор верно утверждает, что объект не принадлежит к рассматриваемому классу (дождь не ожидался и не было)
```

```python
accuracy = (TP+TN)/(TP+TN+FP+FN)
precision = TP/(TP+FP) #процент что дождь был по ожиданиям (оценка доверия модели)
recall = TP/(TP+FN) #процент того, что дождь ожидался, когда был (полнота)
f_score = 2/(1/precision + 1/recall) # среднее гармоническое
false_positive_rate = FP/(FP+TN)

print(accuracy, precision, recall, f_score, false_positive_rate)
```

```python
y_proba = model.predict_proba(X_test)
y_proba
```

```python
y_proba = model.predict_proba(X_test)[:,1]
alpha_list=np.linspace(0,1,100)
tpr = []
fpr = []

for alpha in alpha_list:
  y_pred = y_proba > alpha

  conf_matrix = confusion_matrix(y_test, y_pred)
  TP = conf_matrix[1,1] #классификатор верно отнёс объект к рассматриваемому классу (дождь ожидался и был)
  FN = conf_matrix[1,0] #классификатор неверно утверждает, что объект не принадлежит к рассматриваемому классу (дождь не ожидался, но был)
  FP = conf_matrix[0,1] #классификатор неверно отнёс объект к рассматриваемому классу (дождь ожидался, но не было)
  TN = conf_matrix[0,0] #классификатор верно утверждает, что объект не принадлежит к рассматриваемому классу (дождь не ожидался и не было)

  curr_tpr = TP/(TP+FN)
  curr_fpr = FP/(FP+TN)
  tpr.append(curr_tpr)
  fpr.append(curr_fpr)

tpr = np.array(tpr)
fpr = np.array(fpr)

plt.plot(fpr,tpr)
plt.xlabel('fpr')
plt.ylabel('tpr')
plt.grid()
plt.title('ROC curve')
```

```python
from sklearn.metrics import roc_auc_score
auc = roc_auc_score(y_test, y_proba) # площадь под кривой
auc
```
